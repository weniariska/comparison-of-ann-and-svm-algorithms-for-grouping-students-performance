```python
import pandas as pd
df = pd.read_csv("StudentsPerformanceGrouping.csv")
df.head()
```

```python
# identifikasi missing value

import numpy as np
# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.isnull().sum()
```

```python
# Label encoding -> untuk merubah categorical value menjadi numeric value
from sklearn import preprocessing
le_gen = preprocessing.LabelEncoder()
df['gender'] = le_gen.fit_transform(df['gender'])
le_par = preprocessing.LabelEncoder()
df['parental level of education'] = le_par.fit_transform(df['parental level of education'])
le_lun = preprocessing.LabelEncoder()
df['lunch'] = le_lun.fit_transform(df['lunch'])
le_tes = preprocessing.LabelEncoder()
df['test preparation course'] = le_tes.fit_transform(df['test preparation course'])
df.head()
```

```python
x = df[['gender','parental level of education','lunch','test preparation course','math score','reading score','writing score','average']]
y = df['classgroup']

# normalisai data -> fitur atau 'x' dinormalisasi ke dalam range 0 sampai 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
rescaledX = scaler.fit_transform(x)

# pembagian data menjadi data training dan data testing
x_train = rescaledX[:-10,:]
y_train = y[:-10]
x_test = rescaledX[-10:,:]
y_test = y[-10:]
```

```python
# model
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(activation='relu', solver='lbfgs', max_iter=500, alpha=1e-5, hidden_layer_sizes=(5,3), random_state=0)
clf.fit(x_train, y_train)

# akurasi
from sklearn.metrics import accuracy_score
y_model = clf.predict(x_train)
print("Akurasi validasi : ", accuracy_score(y_train, y_model))

y_prediksi = clf.predict(x_test)
print("\nPerbandingan y_test dan y_prediksi")
print("y_test     : ", np.array(y_test))
print("y_prediksi : ", y_prediksi)
print("\nAkurasi testing : ", accuracy_score(y_test, y_prediksi))

# precision score
from sklearn.metrics import precision_score
print("Precision score : ", precision_score(y_test, y_prediksi,average='micro'))

# recall score
from sklearn.metrics import recall_score
print("Recall score    : ", recall_score(y_test, y_prediksi,average='micro'))
```
